# 01 - Exploração inicial dos dados de gas

Este notebook carrega o dataset consolidado de gas gerado a partir da API Etherscan
(e.g. `data/raw/gas_data_YYYYMMDD_YYYYMMDD.csv`) e apresenta análises descritivas
que podem ser usadas no texto do TCC.

- Estatísticas básicas das métricas (`avgGasPrice_Gwei`, `gasLimit`, `gasUsed`).
- Séries temporais das principais variáveis.
- Correlação entre variáveis.
