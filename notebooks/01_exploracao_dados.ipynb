# 01 - Exploração inicial dos dados de gas

Este notebook carrega o dataset de gas gerado pela API GoldRush (Covalent), por exemplo
`data/gas_data_tcc.csv` (script `fetch_blocks_goldrush`), e apresenta análises descritivas
para o TCC.

- Estatísticas básicas das métricas (`block_height`, `signed_at`, `gas_used`, `gas_limit`, `base_fee`, `tx_count`).
- Séries temporais das principais variáveis.
- Correlação entre variáveis.
